```python
#Age histogram
plt.hist( df['Age'] )

plt.xlabel("Age Groups")
plt.ylabel("Customers number")
plt.title("No. of customer for each age group")
plt.show()


```

```python
#youngest and oldst customer 

youngest = df["Age"].min()
oldest = df["Age"].max()

print("Youngest Customer : ",youngest)
print("Oldest Customer : ",oldest)
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.xlabel("Gender group")
plt.ylabel("Male and Female")
plt.title("No. of customer for each gender")
plt.show()
```

```python
#purchase per category
df.groupby('Category')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
# sort data by review rating and show top 5

df.sort_values('Review Rating',ascending=False, inplace=True)
df.head(5)
```

```python
# show subscription stutus in Pie chart

df['Subscription Status'].value_counts().sort_values(ascending=False).plot(kind='pie')
plt.show()
```

```python
# top 5 colors pie chart
df['Color'].value_counts().head().sort_values(ascending=True).plot(kind='pie')

plt.show()
```

```python
# Pie chart based of gender of customers

plt.pie(df['Gender'].value_counts(), autopct='%f%%',labels=['Male','Female'])
plt.show()
```
