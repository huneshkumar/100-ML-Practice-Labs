```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("final_watch_data.csv")
```

```python
df.head()
```

```python
#split data to X and Y

X= df.drop('Discount Price',axis=1)
y=df['Discount Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2= r2_score(y_test,y_pred)
```

```python
r2
```

```python
#desicison Tree


from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(
    max_depth=5,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)

dt.fit(X_train, y_train)
```

```python
print("Train R2:", dt.score(X_train, y_train))
print("Test R2:", dt.score(X_test, y_test))
```

```python

print("LR R2:", r2_score(y_test, lr.predict(X_test)))
print("DT R2:", r2_score(y_test, dt.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the Random Forest Regressor
rf = RandomForestRegressor(
    n_estimators=100,       # number of trees in the forest
    random_state=42,        # for reproducibility
    max_depth=None,         # maximum depth of each tree, None means nodes are expanded until all leaves are pure
    min_samples_split=2,    # minimum samples required to split an internal node
    min_samples_leaf=1      # minimum samples required at a leaf node
)

# Fit the model to training data
rf.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf.predict(X_train)

# Optional: evaluate using R^2 score
r2_rf = r2_score(y_train, y_pred_rf)
```

```python
print("LR R2:", r2_score(y_test, lr.predict(X_test)))
print("DT R2:", r2_score(y_test, dt.predict(X_test)))
print("Random Forest R2:", r2_rf)
```

```python
score1 = cross_val_score(lr, X, y, cv=5, scoring='r2')
score2 = cross_val_score(dt, X, y, cv=5, scoring='r2')
score3 = cross_val_score(rf, X, y, cv=5, scoring='r2')
print('cross_val_score Linear regression: ', score1, score1.mean(), score1.std())
print('cross_val_score Decision Tree: ', score2, score2.mean(), score2.std())
print('cross_val_score Random Forest: ', score3, score3.mean(), score3.std())
```
