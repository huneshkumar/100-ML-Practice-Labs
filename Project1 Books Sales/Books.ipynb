```python
df=df[df['Publishing Year']>1900]

df.describe()
```

```python
plt.hist( df['Publishing Year'] )
plt.xlabel( "Pulishing Year" )
plt.ylabel( "Frequency" )
plt.title('Distribution of publishing Year')
plt.show()
```

```python
df['genre'].value_counts().plot(kind="bar")
plt.xlabel("Gener")
plt.ylabel("Number of Books")
plt.title("Number of Books in each genre")
plt.show()
```

```python
df.groupby("Author")['Book_average_rating'].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x='genre',y='Book_ratings_count',data=df, hue="genre")
plt.xlabel('Genre')
plt.ylabel('Books Rating Count')
plt.show()
```

```python
plt.scatter(df['sale price'],df['units sold'], hue="sale price")
plt.xlabel(' Sales Price ')
plt.ylabel('No of units sold')
plt.title('Scatter plot for sale price vs units sold')
plt.show()
```
